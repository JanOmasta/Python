```python
# Manually created dataframes. 1 is value for these dataframes.

df_1 = pd.DataFrame(1, index = [1, 3, 6, 4], columns = ["a", "b", 1, 2])
df_2 = pd.DataFrame(1, index = ["i", "j", "a", "b"], columns = ["a", "b", 1, 2])
```

```python
# Access to the dataframe (filtration) using function iloc.

# Filtration of the first and second row of the dataframe.
df_1.iloc[[0, 1]]
```

```python
# Filtration of the first two rows and the first two columns of the dataframe.

df_1.iloc[[0, 1], [0, 1]]
```

```python
df_1.iloc[:2, :2]
```

```python
# Access to the dataframe (filtration) using function loc.
# Here are important values !!! (the name of index - 1, 3, 6, 4).

df_1.loc[:6]
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
# Combined filtration and printing only one column (for this condition, I am interested only in the column SalePrice.).

df.loc[~((df.SalePrice > 175000) & (df.SalePrice < 223500)), "SalePrice"]
```

```python
df.loc[~((df.SalePrice > 175000) & (df.SalePrice < 223500)), ["SalePrice", "YrSold"]]
```

```python
# Alternative code of previous example (without using function loc).

df[~((df.SalePrice > 175000) & (df.SalePrice < 223500))][["SalePrice", "YrSold"]]
```
