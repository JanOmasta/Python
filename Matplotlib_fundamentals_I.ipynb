```
import matplotlib.pyplot as plt

# Import for the chart formating.
from matplotlib.ticker import MaxNLocator
```

```
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49]
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39]
year = [2014, 2015, 2016, 2017, 2018]
```

```
plt.plot(year, apl_price, '--k',                    # k means black color.
         year, ms_price, 'g')
plt.xlabel('Year')
plt.ylabel('Stock price')
# plt.axis([2013, 2019, 20, 180])                   # Modifying the range of horizontal and vertical axises.
plt.show()
```

```
plt.plot(year, apl_price, ':k')
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.show()
```

```
plt.plot(year, apl_price, 'ok')
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.show()
```

```
fig_1 = plt.figure(1, figsize = ([15, 4.8]))         # 1 is a number atribute of a figure.

# The chart 1 and 2 will be added into figure with 20inch width and 4.8inch height.
apple_price = fig_1.add_subplot(121)  # 121 means 1 big row, 2 big columns and 1st chart position on the row.
microsoft_price = fig_1.add_subplot(122)

apple_price.plot(year, apl_price)
apple_price.xaxis.set_major_locator(MaxNLocator(integer = True))

microsoft_price.scatter(year, ms_price)
microsoft_price.xaxis.set_major_locator(MaxNLocator(integer = True))

plt.show()
```

```
fig_1 = plt.figure(1, figsize = ([15, 4.8]))         

# The chart 1 and 2 will be added into figure with 20inch width and 4.8inch height.
apple_price = fig_1.add_subplot(122)  
microsoft_price = fig_1.add_subplot(121)

apple_price.plot(year, apl_price)
microsoft_price.scatter(year, ms_price)
plt.show()
```

```
fig_1 = plt.figure(1, figsize = ([15, 9.6]))         

# Defining the position of the chart in the grid (consisting from 2 rows, 2 columns).
apple_price = fig_1.add_subplot(224)  
microsoft_price = fig_1.add_subplot(221)

apple_price.plot(year, apl_price)
microsoft_price.scatter(year, ms_price)
plt.show()
```

```
# This is shorter way how to write the code above.

fig_2, axes = plt.subplots(2, 2, figsize = (15, 8))
axes[0, 0].scatter(year, ms_price)    # 0 row and 0 position of column.
axes[0, 1].plot(year, apl_price)      # 0 row and 1 position of column.
plt.show()
```
