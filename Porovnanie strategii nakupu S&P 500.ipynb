```python
# (V priemere kazdych 53 obchodnych dni.)

import pandas_datareader
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
S_and_P_500 = pandas_datareader.get_data_yahoo("^GSPC", dt.datetime(2015,6,30))
S_and_P_500
```

```python
S_and_P_500.index = S_and_P_500.index.date
```

```python
S_and_P_500.index
```

```python
S_and_P_500
```

```python
df = S_and_P_500
df
```

```python
# Vypocitaj deltu:
delta = df.Close.diff()

# Vytvor stlpec Delta a prirad do neho premennu delta:
df['Delta'] = delta

# Vymaz cely stlpec Volume, aby mi nezavadzal:
df.drop('Volume', axis = 1, inplace = True)

# Zobraz cely dataframe (jeho hlavicku):
df.head(10)
```

```python
#funkcia clip():

import numpy as np

in_array = [1, 2, 3, 4, 5, 6, 7, 8 ]
print ("Input array : ", in_array)
  
out_array = np.clip(in_array, a_min = 2, a_max = 6)
print ("Output array : ", out_array)
```

```python
delta = S_and_P_500['Close'].diff()
up = delta.clip(lower = 0)
down = -1 * delta.clip(upper = 0)

ema_up = up.ewm(com = 13, adjust = False).mean()
ema_down = down.ewm(com = 13, adjust = False).mean()

rs = ema_up / ema_down
S_and_P_500['RSI'] = 100 - (100/(1 + rs))
S_and_P_500
```

```python
df2 = S_and_P_500.copy()
```

```python
startdate = pd.to_datetime("2016-01-01").date()
enddate = pd.to_datetime("2021-6-10").date()
df3 = df2.loc[startdate:enddate].copy()
df3
```

```python
len(df3)
```

```python
df3['Zaokruh RSI'] = round(df3['RSI'], 2)
df3
```

```python
df3.drop(['High', 'Low', 'Open', 'Adj Close', 'Delta', 'RSI'], axis = 1, inplace = True)
```

```python
df3.rename(columns = {'Zaokruh RSI':'RSI'}, inplace = True)
```

```python
df3
```

```python
df4 = df3.loc[df3['RSI'] <= 30, :].copy()
df4
```
