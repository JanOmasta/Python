```python
import pandas as pd
```

```python
# Defining column names and importing values from Table 1 from excel file.
# Table 1 is caled df_Table_1. (df means Dataframe) 

columns_Table_1 = ['Record_ID_Table_1', 'Vendor_Table_1', 'Model_Name_Table_1']
df_Table_1 = pd.read_excel("Table1.xlsx", header = None, names = columns_Table_1) 
```

```python
df_Table_1
```

```python
# Defining column names and importing values from Table 2 from excel file.
# Table 2 is caled df_Table_2. (df means Dataframe) 

columns_Table_2 = ['Record_ID_Table_2', 'Vendor_Table_2', 'Description_Table_2', 'Model_Name_Table_2']
df_Table_2 = pd.read_excel("Table2.xlsx", header = None, names = columns_Table_2)
```

```python
df_Table_2
```

```python
# Eliminating all duplicate values in column Model Name of Table 2. 
# Created table without duplicates is called df_Table_2_nodups. nodups means no duplicates.

df_Table_2_nodups = df_Table_2.drop_duplicates(subset = ['Model_Name_Table_2'])

df_Table_2_nodups.shape
```

```python
# To merge tables, it is needed to compare values in its tables. 
# To do so, it is needed to change lowercase letters to CAPITAL letters of column Model Name of Table 2 using function upper().

df_Table_2_nodups['Model_Name_Table_2'] = df_Table_2_nodups['Model_Name_Table_2'].str.upper()
```

```python
# Eliminating all duplicate values in column Model Name of Table 1. 
# Created table without duplicates is called df_Table_1_nodups. nodups means no duplicates.

df_Table_1_nodups = df_Table_1.drop_duplicates(subset = ['Model_Name_Table_1'])

df_Table_1_nodups.shape
```

```python
# Merging tables (inner join as from SQL) using columns Model Name from Table 1 and Model Name from Table 2.
# Merged table is called merged_inner.

merged_inner = pd.merge(left = df_Table_1_nodups, 
                        right = df_Table_2_nodups, 
                        left_on = 'Model_Name_Table_1', 
                        right_on = 'Model_Name_Table_2')[['Record_ID_Table_2', 
                                                          'Vendor_Table_2', 
                                                          'Description_Table_2', 
                                                          'Model_Name_Table_2', 
                                                          'Model_Name_Table_1', 
                                                          'Record_ID_Table_1'
                                                         ]]

merged_inner.shape
```

```python
merged_inner.head(25)
```

```python
# Deleting empty rows in created merged table where Model Name (from Table 1) is empty.

merged_inner.dropna(subset = ["Model_Name_Table_1"], inplace = True)

merged_inner.head(5)
```

```python
# Renaming all column names acccording to the column names example in Table 3 in excel file.
# The final table with renamed column names is called df_final_table.

df_final_table = merged_inner.rename(columns = {'Record_ID_Table_2': 'Record ID', 
                                                'Vendor_Table_2' : 'Vendor', 
                                                'Description_Table_2' : 'Description', 
                                                'Model_Name_Table_2' : 'Model Name', 
                                                'Model_Name_Table_1' : 'Model Name (from Table #1)', 
                                                'Record_ID_Table_1' : 'Record ID (from Table #1)'
                                               }
                                    )

df_final_table.head(25)
```

```python
# Writing final table into excel file.
# Create a Pandas Excel writer using XlsxWriter as the engine.

from pandas import ExcelWriter

writer = pd.ExcelWriter('DataMatchingCaseStudy8.xlsx', engine='xlsxwriter')
```

```python
# Writing into excel on sheet called Results.

df_final_table.to_excel(writer, sheet_name = 'Results', startrow = 0, startcol = 0, header = True, index = False)

# Styling of text of final table (font color).

# Get workbook.
workbook = writer.book

# Get Results
worksheet = writer.sheets['Results']

format1 = workbook.add_format()
format2 = workbook.add_format()

format1.set_font_color('#4472C4')
worksheet.set_column('A:A', None, format1)
worksheet.set_column('B:B', None, format1)
worksheet.set_column('C:C', None, format1)
worksheet.set_column('D:D', None, format1)

format2.set_font_color('#ED7D31')
worksheet.set_column('E:E', None, format2)
worksheet.set_column('F:F', None, format2)

writer.save()
```
