```python
# Next chapter.
# Printing all rows where SalePrice is > 181500.

df[df.SalePrice > 181500]
```

```python
# Working with copy of the dataframe. 

# Set for all SalePrice higher than 181500 to value 0.

df[df.SalePrice > 181500].SalePrice = 0
```

```python
df  # The original dataframe is not changed. (See SalePrice column.)
```

```python
# Creating temporal dataframe temp_df.

temp_df = df[df.SalePrice > 181500]

# Changing values in temporal dataframe to 0.

temp_df.SalePrice = 0
```

```python
temp_df  # Temporal dataframe is changed. (See SalePrice column.)
```

```python
# Selecting rows and after that columns (in this order!!) using function loc.
# Set for all SalePrice higher than 181500 to value 0.

df.loc[df.SalePrice > 181500, "SalePrice"] = 0 
df  # Dataframe with changed values.
```

```python
# Chaining conditions. This is a good way.
# : means all columns

df.loc[(df.SalePrice > 100000) & (df.MSSubClass > 30), :]
```
